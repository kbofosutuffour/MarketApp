```python
# Load spaCy model
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])

# Update regex pattern to include hyphens explicitly
pattern = re.compile('[\W_\-]+|')  # Including hyphen explicitly

# Function to clean tokens
def clean_tokens(text):
    # Process the text
    doc = nlp(text)
    # Filter tokens using regex, check if alphabetic, and exclude stopwords
    cleaned_tokens = [
        pattern.sub('', token.lemma_).lower()
        for token in doc
        if pattern.sub('', token.lemma_).isalpha() and not token.is_stop
    ]
    return cleaned_tokens

# Apply function
sample['cleaned_tokens'] = sample['Text'].apply(clean_tokens)

# Display results
print(sample[['Text', 'cleaned_tokens']])
```

```python
def recommend(query, index, product_names, top_k=5):
    query_embedding = model.encode([query], show_progress_bar=False).astype('float32')
    query_embedding = query_embedding.reshape(1, -1)
    distances, indices = index.search(query_embedding, top_k + 1)
    print("Top recommendations:")
    for i, idx in enumerate(indices[0][0:]):
        print(f"{i + 1}: {product_names[idx]} (Score: {1 - distances[0][i]})")

while True:
    user_input = input("Enter a product name to search for recommendations (or type 'exit' to quit): ")
    if user_input.lower() == 'exit':
        break
    recommend(user_input, index, product_names, top_k=5)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

def recommend_items(username):
    # Load data
    data_1 = pd.read_csv('/content/drive/MyDrive/final project/interactionpost/tonychopper_interactions.csv')
    data_2 = pd.read_csv('/content/drive/MyDrive/final project/interactionpost/boruto_interactions.csv')
    data_3 = pd.read_csv('/content/drive/MyDrive/final project/interactionpost/jane_interactions.csv')

    # Combine all data into one DataFrame
    df = pd.concat([data_1, data_2, data_3])

    # Create a user-item matrix where the value is whether the post was liked
    interaction_matrix = df.pivot_table(index='username', columns='post', values='liked_post', fill_value=0)

    # Calculate similarity matrix
    similarity_matrix = cosine_similarity(interaction_matrix)

    # Convert it into a DataFrame, using usernames as both rows and columns
    similarity_df = pd.DataFrame(similarity_matrix, index=interaction_matrix.index, columns=interaction_matrix.index)

    # Check if the user is in the DataFrame's index
    if username in similarity_df.index:
        similar_users = similarity_df[username].sort_values(ascending=False)[1:]  # ignore self-similarity

        # Recommend items
        recommended_items = set()
        for user in similar_users.index:
            # Add items liked by similar users but not yet interacted with by the given username
            recommended_items.update(df[(df['username'] == user) & df['liked_post']]['post'].tolist())

        # Remove items already interacted with by the given username
        recommended_items.difference_update(df[(df['username'] == username)]['post'].tolist())

        # Convert set to list and sort
        recommended_list = sorted(list(recommended_items))

        # Display the top 5 recommendations
        print(f"Top 5 recommended posts for {username}:")
        for item in recommended_list[:5]:
            print(item)
    else:
        print(f"User '{username}' not found. Please check the username.")

recommend_items('TonyChopper')
recommend_items('BorutoUzumaki')
recommend_items('jrhee01')
```
